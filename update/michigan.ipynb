**Downloading file**

Go to this [link](https://data.sca.isr.umich.edu/data-archive/mine.php), select 1978 as the start year and download the "All tables CSV file", and save in the data folder with the name `raw_michigan`

```python
# importing data:
michigan = pd.read_csv("data\\consumer_confidence\\raw_michigan.csv", skiprows=1)
michigan.tail()
```

```python
# date column:
michigan["date"] = pd.to_datetime(michigan["yyyy"].astype(str) + "-" + michigan["Month"].astype(str) + "-01")
michigan = michigan.drop(columns = ["Month", "yyyy"])
```

```python
# melt
michigan = michigan.melt(id_vars="date", var_name="series_id")
michigan = michigan[["series_id","date","value"]].dropna()
michigan.head()
```

Saving File:

```python
michigan.to_parquet("data\\consumer_confidence\\data_michigan_consumer_confidence.parquet")
```
