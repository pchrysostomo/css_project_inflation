## Philly Fed M-PMI

Importing .txt file with all data:

```python
pm_pmi = pd.read_csv("https://www.philadelphiafed.org/-/media/frbp/assets/surveys-and-data/mbos/historical-data/data-series/bos_historyt.txt?la=en&hash=3CE95A52722F29DE526BB7D5D4486A08",
                     delimiter="\t")
```

```python
pm_pmi.head()
```

```python
#
pm_pmi["year"] = pd.to_numeric(pm_pmi["DATE"].str[-2:]) 
pm_pmi.loc[pm_pmi["year"]<67,"year"] = 20
pm_pmi.loc[pm_pmi["year"]>67,"year"] = 19
pm_pmi["year"] = pm_pmi["year"].astype(str)

pm_pmi["date"] = pd.to_datetime("01-"+pm_pmi["DATE"].str[:-2] + pm_pmi["year"] + pm_pmi["DATE"].str[-2:])
pm_pmi = pm_pmi.drop(columns = ["DATE","year"])

pm_pmi.head()
```

```python
pm_pmi = (
    pm_pmi
    .melt(id_vars="date",var_name = "series_id")
    [["series_id","date","value"]]
)

#pm_pmi["date"]= pd.to_datetime("01-" + pm_pmi["date"])
pm_pmi = pm_pmi.sort_values(["series_id","date"]).reset_index(drop=True)

pm_pmi.head(10)
```

Building Dictionary. Documentation available at [here](https://www.philadelphiafed.org/-/media/frbp/assets/surveys-and-data/mbos/readme.txt)

```python
# series_id format: **$%#a
# ** = Item code
# $ = state of the variable (c for current, f for future)
# % = data type (d (%decrease), n (%no change), i (%increase), di(diffusion index)) (only going to select di)
# # = seasonal adjust (s for SA and n for NSA) (only going to select s)
# a = always present

# series codes ---------
# ga: general activity
# no: new orders
# sh: shipments
# dt: delivery time
# iv: inventories 
# uo: unfilled orders
# pp: prices paid
# pr: prices received
# ne: employment
# aw: workhours
# ce: capital expenditures
```
